# Facade

Een facade pattern is een klasse die een eenvoudige `interface` biedt voor een complexe subsysteem dat uit veel onderdelen bestaat. Het biedt mogelijkheid beperkte functionaliteit in vergelijking met het rechtstreeks werken met het subsysteem. Het bevat alleen die functies waar clients echt belang aan hebben.




```typescript
class HomeTheaterFacade{
    private beamer: Beamer = new Beamer();
    private licht: Licht = new Licht();
    private tv: Tv = new Tv();
    private audio: Audio = new Audio();


    public filmKijken(){
        console.log("--- Film kijken ---");
        this.licht.dimmen();
        this.beamer.on();
        this.tv.on();
        this.audio.on();
        console.log("------------------");
    }

    public filmStoppen(){
        console.log("--- Film stoppen ---");
        this.licht.aan();
        this.beamer.off();
        this.tv.off();
        this.audio.off();
        console.log("------------------");
    }
}

class Beamer{
    public on(){
        console.log("Beamer aan");
    }
    public off(){
        console.log("Beamer uit");
    }
}

class Licht{
    public dimmen(){
        console.log("Licht dimmen");
    }
    public aan(){
        console.log("Licht aan");
    }
}

class Tv{
    public on(){
        console.log("Tv aan");
    }
    public off(){
        console.log("Tv uit");
    }
}

class Audio{
    public on(){
        console.log("Audio aan");
    }
    public off(){
        console.log("Audio uit");
    }
}

const theater = new HomeTheaterFacade();
theater.filmKijken();
theater.filmStoppen();
```
