# Decorator

Het decorator pattern is een structural design pattern dat je in staat stelt om dynamisch extra gedrag of verantwoordelijkheden aan een object toe te voegen, zonder oorspronkelijke code te wijzigen.



```typescript
interface Koffie{
    execute()
}

class NormaleKoffie implements Koffie{
    execute(){
        console.log("Normale koffie gemaakt");
    }
}
abstract class KoffieDecorator implements Koffie{
    protected wrappee: Koffie;
    constructor(toWrap: Koffie){
        this.wrappee = toWrap;
    }
    public execute(){
        this.wrappee.execute();
    }
}

class KoffieMetMelk extends KoffieDecorator{
    execute(){
        super.execute();
        this.voegMelkToe();
    }
    voegMelkToe(){
        console.log("Melk toegevoegd");
    }
}

class SuikerDecorator extends KoffieDecorator{
    execute(){
        super.execute();
        this.voegSuikerToe();
    }
    voegSuikerToe(){
        console.log("Suiker toegevoegd");
    }
}

let mijnKoffie = new NormaleKoffie();
mijnKoffie = new KoffieMetMelk(mijnKoffie);
mijnKoffie = new SuikerDecorator(mijnKoffie);
mijnKoffie.execute();
```
