# Factory



```typescript
abstract class StoreFactory{
    public abstract createProduct(): Product;
    public someOperation(): Product{
        let product = this.createProduct();
        console.log(`StoreFactory: Created ${product.getName()}`);
        return product;
    }
}

class StoreAFactory extends StoreFactory{
    public createProduct(): Product{
        return new ProductA();
    }
}

class StoreBFactory extends StoreFactory{
    public createProduct(): Product{
        return new ProductB();
    }
}

interface Product{
    getName(): string;
}

class ProductA implements Product{
    public getName(): string{
        return "Product A";
    }
}

class ProductB implements Product{
    public getName(): string{
        return "Product B";
    }
}

let factoryA = new StoreAFactory();
let factoryB = new StoreBFactory();

let productA = factoryA.someOperation();
let productB = factoryB.someOperation();

console.log(productA.getName());
console.log(productB.getName());
```
