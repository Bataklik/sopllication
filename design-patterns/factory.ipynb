# Factory



```typescript
abstract class Logistics{
    public abstract createTransport(): Product;
    public planDelivery(): Product{
        let product = this.createTransport();
        console.log(`Logistics: Created ${product.constructor.name}`);
        return product;
    }
}

class RoadLogistics extends Logistics{
    public createTransport(): Product{
        return new Truck();
    }
}

class SeaLogistics extends Logistics{
    public createTransport(): Product{
        return new Ship();
    }
}

interface Product{
    deliver(): void;
}

class Ship implements Product{
    public deliver(): void{
        console.log("Ship is being delivered.");
    }
}

class Truck implements Product{
    public deliver(): void{
        console.log("Truck is being delivered.");
    }
}

let factoryA = new RoadLogistics();
let factoryB = new SeaLogistics();

let productA = factoryA.planDelivery();
let productB = factoryB.planDelivery();
console.log("");
productA.deliver();
productB.deliver();
```
