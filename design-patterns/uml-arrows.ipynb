# UML Arrows

- Inheritance | Overerving (generalization): 
    Een open driehoek bij de bovenliggende klassen.
    
    &rarr; Geeft aan dat subklasse specificaties zijn van de bovenliggende klasse.

    

```typescript
// Een Student is een Person
// Een Teacher is een Person
class Person{
    constructor(private name: string, private age: number){}

    public getName(): string {
        return this.name;
    }

    public getAge(): number {
        return this.age;
    }
}

class Student extends Person{
    constructor(name: string,  age: number, private studentId: number){
        super(name, age);
    }
    public getStudentId(): number {
        return this.studentId;
    }
}

class Teacher extends Person{
    constructor(name: string,  age: number, private subject: string){
        super(name, age);
    }
    public getSubject(): string {
        return this.subject;
    }
}


let person = new Person("John", 30);
let student = new Student("Alice", 20, 12345);
let teacher = new Teacher("Bob", 40, "Math");

console.log(person);
console.log(student);
console.log(teacher);
```

- Implementatie | Realisation: Een stippelijn met een open driehoek bij de blauwdrukklasse.

    &rarr; Wanneer een klasse de functionaliteit van een 'blauwdruk'-klasse implementeert en per klasse kan worden aangepast.



```typescript
interface Person{
    getName(): string;
    getAge(): number;
}

class Student implements Person{
    constructor(private name: string, private age: number, private studentId: number){
    }
    public getName(): string {
        return this.name;
    }
    public getAge(): number {
        return this.age;
    }
    public getStudentId(): number {
        return this.studentId;
    }
}

class Teacher implements Person{
    constructor(private name: string, private age: number, private subject: string){
    }
    public getName(): string {
        return this.name;
    }
    public getAge(): number {
        return this.age;
    }
    public getSubject(): string {
        return this.subject;
    }
}


let student = new Student("Alice", 20, 12345);
let teacher = new Teacher("Bob", 40, "Math");

console.log(student);
console.log(teacher);
```

- Afhankelijkheid | Dependency: Een stippel lijn met een open pijl.

    &rarr; Als de ene klasse verandert, verandert de andere klasse ook. De ene klasse is afhankelijk van de andere klasse.

    

```typescript
class Cookie {
    type: string;
    constructor(type: string) {
        this.type = type;
    }
}

class Person {
    eatCookie(cookie: Cookie): boolean {
        console.log(`Eating a delicious ${cookie.type} cookie!`)
        return true;
    }
}

const person = new Person();
const cookie = new Cookie("Chocolate Chip");
person.eatCookie(cookie);
```

- Association | Associatie: Een lijn tussen twee klassen.

    &rarr; Pijlen aan één of beide klassen om aan te geven dat de klassen elkaar 
    kennen.

    
    

```typescript
class Order {
    private orderId: number;
    constructor(orderId: number) {
        this.orderId = orderId;
    }

    getOrderId(): number {
        return this.orderId;
    }
}

class Customer {
    private name: string;
    private orders: Order[] = [];
    constructor(name: string) {
        this.name = name;
    }

    placeOrder(order: Order): void {
        console.log(`${this.name} placed an order with ID: ${order.getOrderId()}`);
        this.orders.push(order);
    }
}

const customer = new Customer("Alice");
const order1 = new Order(101);
const order2 = new Order(102);

customer.placeOrder(order1);
customer.placeOrder(order2);
console.log(customer)
```

- Aggregation | Aggregatie: Een open ruit bij klasse A een speciaal type van associatie dat aangeeft dat klasse B deel is van klasse A.

    

```typescript
class Library {
    private books: Book[] = [];
    addBook(book: Book): void {
        this.books.push(book);
        console.log(`Added book: ${book.getTitle()}`);
    }
    listBooks(): void {
        const titles = this.books.map(book => book.getTitle());
        console.log("Books in the library:", titles.join(", "));
    }
}

class Book {
    private title: string;
    constructor(title: string) {
        this.title = title;
    }
    getTitle(): string {
        return this.title;
    }
}

const library = new Library();
const book1 = new Book("1984");
const book2 = new Book("To Kill a Mockingbird");

library.addBook(book1);
library.addBook(book2);
library.listBooks();
```

- Composition | Compositie: Een gevulde ruit bij klasse A een speciaal type van associatie dat aangeeft dat klasse B deel is van klasse A en niet zonder klasse A kan bestaan.

    

```typescript
class Building {
    private rooms: Room[] = [];
    addRoom(nr: number): void {
        const room = new Room(this, nr);
        this.rooms.push(room);
        console.log(`Added ${room.getNr()} to the building.`);
    }
}

class Room {
    private building: Building;
    private nr: number;
    constructor(building: Building, nr: number) {
        this.building = building;
        this.nr = nr;
    }
    getNr(): number {
        return this.nr;
    }
}

const building = new Building();
building.addRoom(1);
building.addRoom(2);
```
