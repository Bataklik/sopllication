# Singleton

De Singleton pattern deffineert <ins>de statische methode</ins> `getInstance`, die hezelfde examplaar van de eigen klasse retourneert.

De constructor van de `Singleton` moet voor de clientcode *verborgen* blijven. Het aanroepen van de `getInstance` methode moet de enige manier zijn om het object te verkrijgen.



```typescript
class Client{
    constructor(){
        this.singleton = new Singleton();
    }
}

class Singleton{
    private static #instance: Singleton | null;
    private constructor(){
    }

   public static getInstance(){
        if(!this.#instance){
            this.#instance = new Singleton();
        }
        return this.#instance
    }
}

const s1 = Singleton.instance;
const s2 = Singleton.instance;

if (s1 === s2) {
    console.log(
        'Singleton works, both variables contain the same instance.'
    );
} else {
    console.log('Singleton failed, variables contain different instances.');
}
```
