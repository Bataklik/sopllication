# Builder

De builder pattern verdeelt de constructie van een object in een reeks stappen (`buildWalls`, `buildRoof`, `buildWindows`, `buildDoors`). Om een object te maken, voer je een reeks van deze stappen uit.



```typescript
interface Builder {
    buildWalls(): void;
    buildRoof(): void;
    buildWindows(): void;
    buildDoors(): void;
}
class Director {
    constructor(private builder: Builder) {}

    setBuilder(builder: Builder) {
        this.builder = builder;
    }

    build() {
        this.builder.buildWalls();
        this.builder.buildRoof();
        this.builder.buildWindows();
        this.builder.buildDoors();
    }
}

class HouseBuilder implements Builder {
    buildWalls() {
        console.log("Building walls for the house");
    }
    buildRoof() {
        console.log("Building roof for the house");
    }
    buildWindows() {
        console.log("Building windows for the house");
    }
    buildDoors() {
        console.log("Building doors for the house");
    }
}

class IglooBuilder implements Builder {
    buildWalls() {
        console.log("Building walls for the igloo");
    }
    buildRoof() {
        console.log("Building roof for the igloo");
    }
    buildWindows() {
        console.log("Building windows for the igloo");
    }
    buildDoors() {
        console.log("Building doors for the igloo");
    }
}

const houseBuilder = new HouseBuilder();
const iglooBuilder = new IglooBuilder();

const director = new Director(houseBuilder);
director.build();

console.log("... Switching to igloo builder...");
director.setBuilder(iglooBuilder);
director.build();
```
