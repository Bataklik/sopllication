# Strategy

Het strategy pattern stelt voor om een klasse die iets specifieks op veel verschillende manieren uitvoert, te nemen en al deeze algorimen uit te splitsen in afzonderlijke klassen die `stratgieën` worden genoemd.

De oorspronkelijke klasse `context` moet een veld hebben om verwijzing naar een van de strategieën op te slaan. De context klasse geeft het werk door aan de *gekoppelde strategie*.



```typescript
interface SortingStrategy {
    execute(array: number[]): number[];
}

class SortingContext{
    constructor(private strategy:SortingStrategy) {}

    public setStrategy(strategy: SortingStrategy) {
        this.strategy = strategy;
    }

    sort(array: number[]): number[] {
        return this.strategy.execute(array);
    }
}

class BubbleSort implements SortingStrategy {
    execute(array: number[]): number[] {
        console.log("Sorting using Bubble Sort");
        return array;
    }
}

class QuickSort implements SortingStrategy {
    execute(array: number[]): number[] {
        console.log("Sorting using Quick Sort");
        return array;
    }
}

const array = [5, 3, 8, 1, 2];

const sortContext = new SortingContext(new BubbleSort());
sortContext.sort(array);

sortContext.setStrategy(new QuickSort());
sortContext.sort(array);
```
