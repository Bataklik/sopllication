# SOLID

S: Eén taak per klasse/functie.

O: Nieuwe features toevoegen door code toe te voegen, niet door oude code te herschrijven.

L: Subklassen mogen de boel niet laten crashen.

I: Maak kleine, specifieke interfaces in plaats van één gigantische.

D: Koppel je code los via interfaces (Dependency Injection).


## <ins>**S**</ins>ingle Responsibility Principle (SRP)
Een functie of een klasse mag maar één duidelijke taak hebben.

- **Fout**: Een klasse `OrderProcessor` die zowel prijs berekent, de bestelling opslaat en een e-mail verzendt. 

    &rarr; Als de E-mail, prijsberekening of opslag verandert, moet de hele klasse worden aangepast.

- **Goed**: Splitsen van de taken in aparte klassen: `OrderCalculator`, `OrderRepository` en `EmailService`. 

    &rarr; Elke klasse heeft één verantwoordelijkheid en kan onafhankelijk worden aangepast.

## <ins>**O**</ins>pen/Closed Principle (OCP)
Code moet open staan voor uitbreiding, maar gesloten voor aanpassing.
- **Fout**: Een functie `calculateDiscount()`, als de klant *PREMIUM* is krijgt hij 10% korting. Nu komt er een neiuwe optie *VIP* bij met IF/ELSE in de bestaande code.

    &rarr; Hier past je de bestaande code aan, wat kan leiden tot fouten en regressies.
    
- **Goed**: Een interface `DiscountStrategy` en maakt een klasse `PremiumDiscount` en `VipDiscount`. 
    
    &rarr; Hiermee breidt je de code uit met nieuwe klasse zonder de oude werkende code aan te passen.

### <ins>**L**</ins>iskov Substitution Principle (LSP)
Een subklasse moet kunnen gebruikt worden ipv. de basisklasse zonder dat de app crasht.
- **Fout**: Een klasse `Bird` met een methode `fly()`. Een subklasse `Penguin` die de methode `fly()` niet kan implementeren. 

    &rarr; Als je een `Penguin` object gebruikt waar een `Bird` object verwacht wordt, zal de app crashen.
- **Goed**: Niet elke vogel vliegt, dus opsplitsen in `Bird` en `FlyingBird`. `Penguin` erft alleen van `Bird`.

    &rarr; Hierdoor kan een `Penguin` object gebruikt worden waar een `Bird` object verwacht wordt zonder dat de app crasht.

## <ins>**I**</ins>nterface Segregation Principle (ISP)
Klassen mogen niet gedwongen worden om interfaces te implementeren die ze niet gebruiken.
- **Fout**: Een interface `SmartDevice` met `print()`, `fax()` en `scan()`. Als er een klasse `EpsonPrinter` is die alleen kan printen, moeten de methodes `fax()` en `scan()` ook geïmplementeerd worden, wat leidt tot lege methodes of uitzonderingen.

    &rarr; Dit maakt de code moeilijker te onderhouden en kan leiden tot fouten.

- **Goed**: Maak kleinere interfaces: `Printer`, `Scanner` en `Fax`. De `EpsonPrinter` implementeert alleen de `Printer` interface.

    &rarr; Hierdoor hoeft de klasse alleen de methodes te implementeren die relevant zijn voor zijn functionaliteit.

## <ins>**D**</ins>ependency Inversion Principle (DIP)
Code moet afhankelijk zijn van abstracties (interfaces), niet van concrete implementaties.
- **Fout**: `OrderProcessor` maakt rechtstreeks `const db = new MySQLDatabase()` aan. Als je wilt veranderen naar een andere `DB`, moet je de code aanpassen.

    &rarr; Dit maakt de code minder flexibel en moeilijker te testen.

- **Goed**: `OrderProcessor`vraagt een `DB` (interface), dat `mysql` of `MongoDB` is, maakt voor processor niet uit.

    &rarr; Hierdoor kan de implementatie van de database worden gewijzigd zonder dat de `OrderProcessor` hoeft te worden aangepast.

```typescript
interface Engine {}
class V8Engine implements Engine {}
class ElectricEngine implements Engine {}

class Car{
    private engine = new V8Engine();
}
class DICar{
    private engine: Engine;
    constructor(engine: Engine){
        this.engine = engine;
    }
}

const v8Car = new DICar(new V8Engine());
const electricCar = new DICar(new ElectricEngine());
const defaultCar = new Car();
console.log(v8Car, electricCar, defaultCar);
```
