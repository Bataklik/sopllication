# Clean code

## A. Betekenisvolle namen
Namen moeten vertellen waarom iets bestaat, wat het doet en hoe het wordt gebruikt.
- **Fout**:
  - `let d = 3`, dagen sinds laatste update.
  - `function gegevens(x){...}`
- **Goed**:
  - `let daysSinceLastUpdate = 3`
  - `function fetchUserData(userId){...}`

## B. Functies doen één iets (Single Responsibility)
Een functie moet klein zijn en één taak hebben. **Vuistregel**: Naam van de functie kunnen uitspreken <ins>zonder</ins> `EN` te gebruiken.
- **Fout**:
  - `function validateFormAndSaveToDatabaseAndRedirectUser()`
- **Goed**:
  - `function validateForm()`
  - `function saveToDatabase()`
  - `function redirectUser()`

## C. Vermijd *Magic numbers* en *Comments die code praten*
- Magic numbers: gebruik geen losse getallen in je code waarvan niemand de betekenis kent.
  - **Fout**:
    - `if (user.age > 18) {...}`
  - **Goed**:
    - `const LEGAL_AGE = 18;`
    - `if (user.age > LEGAL_AGE) {...}`
- Comments: 
  - **Fout**:
    - `// Check if user is adult`
    - `if (user.age > LEGAL_AGE) {...}`
  - **Goed**:
    - `if (user.age > LEGAL_AGE) {...}`
